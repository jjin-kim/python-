## 데이터 타입

3.execute  
**cur.execute('SQL')**

```python
cur.execute('SQL')
```

```python
cur.execute("create table people (name_last,age)")
```

```python
who="Yeltsin"
age=72

#this is the qmark style:
cur.execute("insert into people values (?,?)",(who,age))
```

```python
#this is the named style:
cur.execute("select * from people where name_last=:who and age=:age",{"who":who,"age":age})

print(cur.fetchone()) #fetchone()는 cur이 물고있는 데이터를 하나씩하나씩 꺼내서 보여줌
```

4.executemany  
**cur.execuemany('SQL',params)**

```python
sql="insert into people values (?,?)"
curData=[('A',1),('B',2),('C',3)]

cur.executemany(sql,curData)
```

5.executescript  
**cur.executescript('''SQL; SQL;....''')**

```python
cur.executescript("""
    create table person(
        first_name text primary key,
        last_name text not null
    );
        
    insert into person values ('name','kim');
""")
```

```python
cur.execute('select * from person')
print(cur.fetchall())
```

## DDL 

1.CREATE 

```python
conn=sqlite3.connect('create.db')
print("Opened database successfully") #이것은 안써도 되는데 명시할려고 쓴거다.
```

```python
conn.execute('''
    CREATE TABLE COMPANY(
        ID INT PRIMARY KEY   NOT NULL,
        NAME           TEXT  NOT NULL,
        AGE            INT   NOT NULL,
        ADDRESS        CHAR(50),
        SALARY         REAL);
    ''')
print("Table created successfully")
```

2.DROP

```python
conn.execute('drop table company')
```

## DDL 

1.INSERT

```python
conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)\
              VALUES (1,'paul',32,'california',20000.00)");

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)\
              VALUES (:id,:name,:age,:address,:salary)",
              {'id':2,'name':'Allen','age':25,'address':'Texas','salary':15000.00});

data=[(3,'Teddy',23,'Norway',200000.00),
      (4, 'Mark', 25, 'Rich-Mond',65000.00)]

conn.executemany("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)\
                  VALUES(?,?,?,?,?)",data);

conn.executescript("""
    INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)
    VALUES (5,'Mark1',25,'Rich-omnd',65000.00);
    
    INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)
    VALUES (6,'Mark2',25,'Rich-omnd',65000.00);
        """);

conn.commit()
```

2.SELECT

```python
cursor=conn.execute("SELECT id, name, address, salary from COMPANY")

for row in cursor:
    print("ID=",row[0])
    print("NAME=",row[1])
    print("ADDRESS=",row[2])
    print("SALARY=",row[3],end='\n\n')
    
print("Operation done successfully")    
```

3.UPDATE

```python
cid=1

conn.execute("UPDATE COMPANY set SALARY = 25000.00 where ID = :id",{'id':cid})
conn.commit()

print("Total number of rows update :", conn.total_changes)

cursor= conn.execute("SELECT id, name, address, salary from COMPANY")

for row in cursor:
    print("ID=",row[0])
    print("NAME=",row[1])
    print("ADDRESS=",row[2])
    print("SALARY=",row[3],end='\n')
        
```

4.DELETE

```python
conn.execute("DELETE from COMPANY where ID=3;")

print("Total number of rows deleted:", conn.total_changes)

cursor = conn.execute("SELECT id, name, address, salary from COMPANY")

for row in cursor:
    print("ID=",row[0])
    print("NAME=",row[1])
    print("ADDRESS=",row[2])
    print("SALARY=",row[3],end='\n')

conn.commit()
```

# 예제

```python
import sqlite3 as sql

con=None

try:
    con = sql.connect('practice.db')
    cur = con.cursor()
    cur.execute('SELECT SQLITE_VERSION()')
    data = cur.fetchone()
    print("SQLite version:",data)
    
except Exception as e:
    print("Error",e)
    
finally:
    if con:
        con.close() #모든 데이터 입출력시 close()문 필수!!.
```

```python
con=sql.connect('practice.db') #아까 finally구문에서 닫았으므로 다시 열어줌.

with con:
    cur = con.cursor()
    cur.execute("CREATE TABLE cars(id INT, Name TEXT, Price INT)")
    cur.execute("INSERT INTO cars VALUES(1,'Audi',52345)")
    cur.execute("INSERT INTO cars VALUES(2,'Mercedes',54345)")
    cur.execute("INSERT INTO cars VALUES(3,'Skoda',50345)")
    cur.execute("INSERT INTO cars VALUES(4,'Volvo',52305)")
    cur.execute("INSERT INTO cars VALUES(5,'Bentley',52349)")
    cur.execute("INSERT INTO cars VALUES(6,'Citroen',92345)")
    cur.execute("INSERT INTO cars VALUES(7,'Hummer',32345)")
    cur.execute("INSERT INTO cars VALUES(8,'Volkswagen',52145)")
```

```python
cars=[
    (1,'Audi',52345),
    (2,'Mercedes',54345),
    (3,'Skoda',50345),
    (4,'Volvo',52305),
    (5,'Bentley',52349),
    (6,'Citroen',92345),
    (7,'Hummer',32345),
    (8,'Volkswagen',52145)
]

con=sql.connect('practice.db')

with con:
    cur=con.cursor()
    cur.execute("DROP TABLE IF EXISTS Cars")
    cur.execute("CREATE TABLE Cars(Id INT PRIMARY KEY,Name TEXT, Price INT)")
    cur.executemany("INSERT INTO Cars VALUES(?,?,?)",cars)
```

```python
try:
    con = sql.connect('practice.db')
    
    cur = con.cursor()
    
    cur.executescript("""
        DROP TABLE IF EXISTS Cars;
        CREATE TABLE cars(id INT, Name TEXT, Price INT);
        INSERT INTO cars VALUES(1,'Audi',52345);
        INSERT INTO cars VALUES(2,'Mercedes',54345);
        INSERT INTO cars VALUES(3,'Skoda',50345);
        INSERT INTO cars VALUES(4,'Volvo',52305);
        INSERT INTO cars VALUES(5,'Bentley',52349);
        INSERT INTO cars VALUES(6,'Citroen',92345);
        INSERT INTO cars VALUES(7,'Hummer',32345);
        INSERT INTO cars VALUES(8,'Volkswagen',52145);
        """)
    con.commit()
    
except Exception as e:
    if con:
        con.rollback()
        
    print("Error:",e)
    
finally:
    if con:
        con.close()
```

- lastrowid  
-> 마지막 값을 보여줌

```python
con = sql.connect('Friends.db') #메모리상에 데이터 연결

with con:
    cur = con.cursor()
    cur.executescript("""
        DROP TABLE IF EXISTS Friends;
        CREATE TABLE Friends(Id INTEGER PRIMARY KEY, Name TEXT);
        INSERT INTO Friends VALUES(1,'Tom');
        INSERT INTO Friends VALUES(2,'jin');
        INSERT INTO Friends VALUES(3,'kim');
        INSERT INTO Friends VALUES(4,'moon');
        """)
    lid=cur.lastrowid
    print("The last Id of the inserted row is",lid)
```

- fetchone

```python
con = sql.connect("practice.db")



with con:
    cur=con.cursor()
    cur.execute("SELECT * FROM Cars")
    
    while True:
        row=cur.fetchone()
        
        if row==None:
            break
            
        print(row[0],row[1],row[2])
```

- fetchall

```python
con=sql.connect('practice.db')

with con:
    cur=con.cursor()
    
    cur.execute("SELECT * FROM Cars")
    rows= cur.fetchall()
    
    for row in rows:
        print(row[0],row[1],row[2])
```

- UPDATE

```python
uid=1
uprice=400000

con=sql.connect('practice.db')

with con:
    cur = con.cursor()
    
    cur.execute("UPDATE Cars SET price=? WHERE Id=?",(uprice,uid))
    
    print("Number of rows updated:",cur.rowcount)
```

```python
uid=4

con=sql.connect('practice.db')

with con:
    cur = con.cursor()
    
    cur.execute("SELECT Name, Price FROM Cars WHERE Id=Id",{"Id":uid})
    
    row=cur.fetchone()
    print(row[0],row[1])
```

```python
conn= sqlite3.connect("music.db")


with conn:
    cur=conn.cursor()
    cur.executescript("""
        CREATE TABLE Artist(
            id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
            name TEXT);
        CREATE TABLE Genre(
            id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
            name TEXT);
        CREATE TABLE Album(
            id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
            title TEXT,
            artist_id INTEGER);
        CREATE TABLE Track(
            id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
            title TEXT,
            length INTEGER,
            rating INTEGER,
            count INTEGER,
            album_id INTEGER,
            genre_id INTEGER);    
         INSERT INTO Artist(name) values('Led Zepplin');
         INSERT INTO Artist(name) values('AC/DC');
         INSERT INTO Genre(name) values('Rock');
         INSERT INTO Genre(name) values('Metal');
         INSERT INTO Album(title,artist_id) values('Who Made Who',2);
         INSERT INTO Album(title,artist_id) values('Iv',1);
         INSERT INTO Track(title,rating,length,count,album_id,genre_id) values('Black Dog',5,297,0,2,1);
         INSERT INTO Track(title,rating,length,count,album_id,genre_id) values('Stairway',5,482,0,2,1);
         INSERT INTO Track(title,rating,length,count,album_id,genre_id) values('About to Rock',5,313,0,2,1);
         INSERT INTO Track(title,rating,length,count,album_id,genre_id) values('Who Made Who',5,207,0,2,1);
        """)
```

- Join

```python
cur.execute('SELECT Album.title,Artist.name FROM Album JOIN Artist ON Album.artist_id=Artist.id')
```

```python
for row in cur.fetchall():
        print(row)
```

```python
cur.execute('SELECT a.title,b.name from Track as a JOIN Genre as b on a.genre_id=b.id')
```

```python
for row in cur.fetchall():
        print(row)
```

```python
cur.execute('select Track.title, Artist.name, Album.title, Genre.name\
             from Track\
             join Artist join Album join Genre on\
             Track.album_id = Album.id\
             and\
             Track.genre_id = Genre.id\
             and\
             Album.artist_id = Artist.id
            ')
for row in cur.fetchall():
        print(row)
```
